#### Boolean Columns

```python
# Creaate a new column that is true if row is RB
pg['is_a_rb'] = (pg['pos'] == 'RB')
```

```python
pg.sample(5)
```

```python
pg[['player_name', 'is_a_rb']].sample(5)
```

```python
# Combining logical operations '|' and '&'
pg['is_a_rb_or_wr'] = (pg['pos'] == 'RB') | (pg['pos'] == 'WR')
pg[['player_name', 'is_a_rb_or_wr', 'pos']].sample(5)
```

```python
pg['good_rb_game'] = ((pg['pos'] == 'RB') & (pg['rush_yards'] >= 100))
pg.sample(5)
```

```python
pg['is_not_a_rb_or_wr'] = ~((pg['pos'] == 'RB') | (pg['pos'] == "WR"))
pg.sample(5)
```

```python
# Can work with multiple columns at once

(pg[['rush_yards', 'rec_yards']] >= 100).sample(5)
```

#### Applying Functions to Columns

```python
def is_skill(pos):
    """
    Takes some string named pos ('QB', 'K', 'RT' etc) and checks whether it's a skill postion (RB, WR, TE).
    """
    return pos in ['RB', 'WR', 'TE']
```

```python
pg['is_skill'] = pg['pos'].apply(is_skill)
```

```python
pg[['player_name', 'is_skill']].sample(5)
```

```python
# lambda function is useful here
pg['is_skill_alternate'] = pg['pos'].apply(lambda x: x in ['RB', 'WR', 'TE'])
pg.sample(5)
```

#### Dropping columns

```python
pg.drop('is_skill_alternate', axis = 1, inplace = True)
```

#### Renaming Columns

```python
# renaming all columns in pg to be uppercase

pg.columns = [x.upper() for x in pg.columns]
pg.head()
```

```python
# Changing it back since convention is lowercase
pg.columns = [x.lower() for x in pg.columns]
```

```python
# Another way to rename columns
pg.rename(columns = {'interceptions' :'ints'}, inplace =True)
```

#### Changing column types

```python
# notice the gameid column had a string of the date. Let's grab the date
pg['gameid'].head()
```

```python
gameid = '2017090700'

year = gameid[0:4]
month = gameid[4:6]
day = gameid[6:8]

print(month, day,year)
```

```python
# trying string method on a column
pg['month'] = pg['gameid'].str[4:6]
```
