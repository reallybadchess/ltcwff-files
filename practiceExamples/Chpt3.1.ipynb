#### Copies and the inplace argument

```python
# our index didn't save. SAD!
adp.head()
```

```python
# to make the change permanent we can either set inplace argument to True
adp.set_index('player_id', inplace = True)
```

```python
adp.head()
```

```python
# Or we can override the data frame with a new variable
adp = adp.set_index('player_id')
```

```python
adp
```

```python
# creating an RB subset dataframe
adp_rbs = adp.loc[adp['position'] == 'RB', ['name','position','team']]
adp_rbs.head()
```

```python
# another df to sort them by name
adp_rbs.sort_values('name', inplace = True)
adp_rbs.head()
```

```python
# what if we want to add a new column from the original DF?
adp_rbs['times_drafted'] = adp['times_drafted']
adp_rbs.head()
```

### Outputting Data

```python
adp_rbs.to_csv(path.join(dataDir, 'adp_rbs.csv'))
```
