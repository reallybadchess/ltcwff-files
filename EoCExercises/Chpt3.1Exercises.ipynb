```python
# Importing libraries
import pandas as pd
from os import path
```

```python
# 3.0.1 - Load the ADP data into a DF named adp.

# path variable
dataDir = '/Users/simmsjn/Documents/GitHub/ltcwff-files/data'

# creating the ADP df
adp = pd.read_csv(path.join(dataDir, 'adp_2017.csv'))
```

```python
# 3.0.2 - Use the adp df to create another df, adp50, that is the top 50 (by ADP) players

adp50 = adp.sort_values('adp').head(50)
adp50
```

```python
# 3.0.5 
# A) Make a new df, adpSimple, with just the columns 'name', 'position', 'adp', in that order

adpSimple = adp[['name','position', 'adp']]
adpSimple.head()
```

```python
# 3.0.5
# B) rearrange addpSImple so the orderr is 'position', 'name', 'adp'.

adpSimple = adp[['position', 'name', 'adp']]
adpSimple.head()
```

```python
# 3.0.5
# C) Using the original adp DF, add the 'team' column to adpSimple

adpSimple['team'] = adp['team']
adpSimple.head()

```
