```python
# importing libraries

import numpy as np
import pandas as pd
from os import path
```

```python
# 3.1.1 - Load the player game data into a DF named pg
# creating the path variable

dataDir = '/Users/simmsjn/Documents/GitHub/ltcwff-files/data'

# creating the DF

pg = pd.read_csv(path.join(dataDir, 'player_game_2017_sample.csv'))
pg.head()
```

```python
# 3.1.2 - Add a column to pg, 'rec_pts_ppr', that give total receiving fantasy points, where the scoreing system
# is 1 point for every 10 yeards receving, 6 points per receiving TD and 1 point per reception

pg['rec_pts_ppr'] = (pg['rec_yards']*0.1 + pg['rec_tds']*6 + pg['receptions'])
```

```python
# 3.2.3 - add a column of the form: 'T.Brady is the NE QB'

pg['player_desc'] = pg['player_name'] + ' is the ' + pg['team'] + ' ' + pg['pos']
pg.head()
```

```python
# 3.1.4 - Add a boolean column to pg 'is_possesion_rec' indicating whether a players air yards were greater
# than his total yards after the catch
# raw_yac: raw yards after catch
pg['is_possesion_rec'] = (pg['caught_airyards'] == pg['raw_yac']) 
pg.head()
```

```python
# 3.1.5 - add a column 'len_last_name' that gives the length of the players last name

pg['last_name'] = pg['player_name'].apply(lambda x: len(x.split('.')[1]))
pg.head()
```

```python
#3.1.6 - 'gameid' is a numberic(int) bu its not meant for math. change it to a string

pg['gameid'] = pg['gameid'].astype(str)
```

```python
#3.1.7 - replace all '_' with ' ' in the column names

pg.columns = [x.replace('_', ' ') for x in pg.columns]
pg.head()
```

```python
# 3.1.7 - Changing it back 

pg.columns = [x.replace(' ', '_') for x in pg.columns]
pg.head()
```

```python
#3.1.8 
# A) Make a new column 'rush_td_percentage' indicating the percentage of a players carries that went for TDs.

pg['rush_td_percentage'] = pg['rush_yards']/pg['carries']
pg['rush_td_percentage'].head()
```

```python
#3.1.8
# B) replace missing values with -99

pg['rush_td_percentage'].fillna(-99, inplace = True)
pg['rush_td_percentage'].head()
```

```python
# 3.1.9 - drop the column 'rush_td_percentage' and confirm that it worked

pg.drop('rush_td_percentage', axis = 1, inplace = True)
```
