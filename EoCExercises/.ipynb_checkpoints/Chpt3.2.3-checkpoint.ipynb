```python
# 3.3.1 - Load the ADP data

# importing libraries
import pandas as pd
import numpy as np
from os import path

# path to data
dataDir = '/Users/simmsjn/Documents/GitHub/ltcwff-files/data'

# loading the DF

adp = pd.read_csv(path.join(dataDir, 'adp_2017.csv'))
```

3.3.2 

    Make a smaller DF with just the Dallas Cowboys players and onlt the columns name, adp, position 
    Do it two ways: using loc and query

```python
# A) Using loc
adp_cb1 = adp.loc[adp['team'] == 'DAL', ['name', 'position', 'adp']]
adp_cb1.head()
```

```python
# B) Using query

adp.query("team == 'DAL'")[['name', 'position', 'adp']].head()
```

```python
# 3.3.3 - Make a DF adp_no_cb that is everyone except DAL players add the 'team column to it'

adp_not_cb = adp.loc[adp['team'] != 'DAL', ['name', 'position', 'adp', 'team']]
adp_not_cb.sample(5)
```

```python
# 3.3.4 - Are there any duplicates by last name and position

# Create last name column
adp['lastName'] = adp['name'].apply(lambda x: x.split(' ')[1])

# Checking for duplicates in lastName and position
adp.duplicated(['lastName', 'position']).any()

# gives True therefore there are duplicates
```

```python
# 3.3.5 - Add a new column called adp_description with the values:

        
import numpy as np

adp['adpDescription'] = np.nan
adp.loc[adp['adp'] < 40, 'adpDescription'] = 'stud'
adp.loc[adp['adp'] > 120, 'adpDescription'] = 'scrub'
adp[['adp', 'adpDescription']].sample(5)
```

```python
# 3.3.6 - make a new DF with only observations for which 'adpDescriptions' is missing
# Do this both with loc and query syntax.
# Call them adp_no_desc1 and apd_no_desc2
```

```python
# a) With loc

adp_no_desc1 = adp.loc[adp['adpDescription'].isnull()]
adp_no_desc1.head()
```

```python
adp_no_desc2 = adp.query("adp_description.isnull()")
```
