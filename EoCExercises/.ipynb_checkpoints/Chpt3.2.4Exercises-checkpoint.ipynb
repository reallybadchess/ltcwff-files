
#### 3.4.1 How does shifting granularites affect the amount of information in your data?

Granularity allows for us to to look at the data at different levels. Choosing to look at player stats per
week instead of per game or per play. We can gain or lose information depending on the granularity we choose.

#### C) Figure out everyone's ypc

```python
(pbp.query("play_type == 'run'").groupby(['game_id', 'rusher_player_name'])['yards_gained'].mean())


```

```python
pbp['lte_0_yards'] = pbp['yards_gained'] <= 0

(pbp
.query("play_type == 'run'")
.groupby(['game_id', 'rusher_player_name'])['lte_0_yards'].mean())
```
